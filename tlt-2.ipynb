```python
import re
import time
import requests
from bs4 import BeautifulSoup as bs
import datetime as dt
import json
import ast  #轉換成json需要套件
import string

# 標題,類別,內文,關鍵詞,新聞連結網址,日期(格式yyyyMMdd)


def wr_all(url):  #方法 大部分的網頁版型
    res = requests.get(url)
    res.encoding = "UTF-8"
    soup = bs(res.text)
    essay = soup.select('#newstext p , h4')
    tital = soup.select('h1 ')[0].text
    date =  soup.select('#newstext span ')[0].text
    category = soup.select('.guide  a ')[1].text
    keyword= soup.select('.con_keyword')
    date_ture = string.replace(date, '-', '')[0:8]
    page=""                          #給一個字串,存放內文用           
    for div3 in essay[0:(len(essay)-1)]:
        page += div3.text       #加入內文
    key=""                          #給一個字串,存放關鍵字用 
    for div4 in essay[1:(len(keyword))]:
        key += div4.text

    return tital,date_ture,category, page,key
        
def wr_sport(url):  #方法 運動類的網頁版型
    res = requests.get(url)
    res.encoding = "UTF-8"
    soup = bs(res.text)    
    essay = soup.select('h4 , p')
    tital = soup.select('.Btitle ')[0].text
    date =  soup.select('.news_content .date ')[0].text
    category = "體育"
    date_ture = string.replace(date, '-', '')[0:8]

    page=""                          #給一個字串,存放內文用           
    for div3 in essay[0:(len(essay))]:
        page += div3.text       #加入內文  
    return tital,date_ture,category, page
     
def wr_ent(url):  #方法 娛樂類的網頁版型
    res = requests.get(url)
    res.encoding = "UTF-8"
    soup = bs(res.text)
    essay = soup.select('p')
    tital = soup.select('.news_content h1 ')[0].text
    date =  soup.select('.news_content .date')[0].text
    date_ture = string.replace(date, '-', '')[0:8]
    category = "娛樂"
    page=""                          #給一個字串,存放內文用           
    for div3 in essay[0:(len(essay))]:
        page += div3.text       #加入內文   
    return tital,date_ture,category, page

def wr_local(url): #方法 地方類的網頁版型
    res = requests.get(url)
    res.encoding = "UTF-8"
    soup = bs(res.text)
    essay = soup.select('#newstext p , h4')
    tital = soup.select('h1 ')[0].text
    date =  soup.select('#newstext span ')[0].text
    category = "地方"
    date_ture = string.replace(date, '-', '')[0:8]
    page=""                          #給一個字串,存放內文用           
    for div3 in essay[0:(len(essay)-1)]:
        page += div3.text       #加入內文
    return tital,date_ture,category, page
    
def wr_talk(url): # 方法 言論類的網頁版型
    res = requests.get(url)
    res.encoding = "UTF-8"
    soup = bs(res.text)
    essay = soup.select('.cont p')
    tital = soup.select('h2 ')[0].text
    date =  soup.select('.writer span')[0].text
    date_ture = string.replace(date, '-', '')[0:8]
    category = "言論"
    page=""                          #給一個字串,存放內文用           
    for div3 in essay[0:(len(essay))]:
        page += div3.text       #加入內文    
    return tital,date_ture,category, page
        
    
all_json = ""

for i in range(1,3): # 1-2月
    with open('C:/Users/BIG DATA/Desktop/python/essay/20160{}.txt'.format(i), 'r') as f:
        for line in f.readlines():
            url = line.strip()
            m = re.match('.*news/(\w+)/paper/(\w+)' ,url)  #正規表達法 切出m.group(1)類別 與 m.group(2)網頁最後的數字(用來當檔名)
            print m.group(1)
            if m.group(1) == 'sports': #判斷是否為體育類
                x=  wr_sport(url) #使用方法
                data ={'tital' :x[0],'category' : x[2],'essay' : x[3],'url':url,'date':x[1]}
                all_json+=str(data)+','
                time.sleep(0.5)
                
            elif m.group(1) == 'entertainment':#判斷是否為娛樂類
                x= wr_ent(url) #使用方法
                data ={'tital' :x[0],'category' : x[2],'essay' : x[3],'url':url,'date':x[1]}
                all_json+=str(data)+','
                time.sleep(0.5)


            elif m.group(1) =='local': #判斷是否為地方類
                x= wr_local(url) #使用方法
                data ={'tital' :x[0],'category' : x[2],'essay' : x[3],'url':url,'date':x[1]}
                all_json+=str(data)+','
                time.sleep(0.5)

            elif m.group(1) =='opinion': #判斷是否為言論類
                x= wr_talk(url) #使用方法
                data ={'tital' :x[0],'category' : x[2],'essay' : x[3],'url':url,'date':x[1]}
                all_json+=str(data)+','
                time.sleep(0.5)
 
            else:
                x= wr_all(url) #使用方法
                data ={'tital' :x[0],'category' : x[2],'essay' : x[3],'keyword':x[4],'url':url,'date':x[1]}
                all_json+=str(data)+','
                time.sleep(0.5)





with open("new12.json", "w") as f:
    json.dump(all_json, f, indent=4, ensure_ascii=True)
    f.close()
    
```
