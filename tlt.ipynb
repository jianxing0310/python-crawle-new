#  一天頭版新聞

```python
import requests
from bs4 import BeautifulSoup as bs
import time

url = 'http://news.ltn.com.tw/newspaper/focus/20160310'
res = requests.get(url)
soup = bs(res.text)
title = soup.select('.picword')
title_num = len(soup.select('.picword')) #計算有幾篇文章(每天頭條新聞數量不同)
Times = 'http://news.ltn.com.tw'

for i in xrange(0 , title_num):   
    url1=Times+title[i]['href']   #文章網址
    res1 = requests.get(url1)
    soul1 = bs(res1.text)
    essay = soul1.select('#newstext p , h4 ')
    tital = soul1.select('  h1 , .con_keyword , #newstext span ,.guide  a ')
    page=""                          #給一個字串,存放內文用           
    for div3 in essay[0:(len(essay))]:
        page += div3.text       #加入內文
    #print  type(tital)
    print "成功抓取一篇文章"
    name =  title[i]['href'].split('paper/')[1]
    f = open('C:/Users/BIG DATA/Desktop/python/essay/{}.txt'.format(name),'w')
    iii_num = len(tital)
    f.write( page .encode('utf-8')+ '\n')        #內文
    for iii in xrange (0, iii_num):
        f.write( tital[iii].text .encode('utf-8') + '\n') #標題 類別 時間 相關關鍵字          
    f.close()
    time.sleep(3)
```
